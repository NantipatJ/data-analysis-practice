```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
sales_2017 = df[ df["Order Date"].dt.year == 2017 ]

print( sales_2017["Sales"].sum() )
print( sales_2017["Sales"].mean() )
print( sales_2017["Sales"].std() )
```

```python
# TODO 06 - which Segment has the highest profit in 2018
order_2018 = df[ df["Order Date"].dt.year == 2018 ]

profit_by_seg = order_2018.groupby("Segment")["Profit"].sum()

highest_pro_2018 = profit_by_seg.sort_values(ascending = False)
highest_pro_2018
# Consumer segment make the highest profit in 2018
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
df_2019 = df[ (df["Order Date"] >= "2019-04-15") & (df["Order Date"] <= "2019-12-31") ]

state_sales = df_2019.groupby("State")["Sales"].sum()

least_state_sales = state_sales.sort_values(ascending = True).reset_index()
least_state_sales
# Top 5 5 have the least sales is New Hamshipre, New Maxixo, District of Columbia, Louisiana, and South Carolina
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
target_rows = df[ (df["Order Date"].dt.year == 2019) & (df["Region"].isin( ["West", "Central"]) ) ]

west_central_sales = target_rows["Sales"].sum()

total_sales_2019 = df[ df["Order Date"].dt.year == 2019]["Sales"].sum()

proportion = (west_central_sales / total_sales_2019) * 100

clean_proportion = round(proportion, 2)
clean_proportion
# Proportion of total sales in West + Central is 54.97%
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
df_19_20 = df[ df["Order Date"].dt.year.isin( [2019, 2020] ) ]

products =df_19_20.groupby("Product Name").agg(
    order_count = ("Order ID", "count"),
    total_sales = ("Sales", "sum")
)
# Top 10 Products
top_10_products = products.sort_values("order_count", ascending = False)
top_10_products

# Total Sales
top_10_sales = products.sort_values("total_sales", ascending = False)
top_10_sales
```

```python
# TODO Bonus - use np.where() to create new column in dataframe to help you answer your own questions
import numpy as np

product_check = df.groupby("Product Name").agg(
    order = ("Order ID", "count"),
    profit = ("Profit", "sum")
).reset_index()

avg_orders = product_check["order"].mean()
avg_profit = product_check["profit"].mean()

popular_check = product_check["is_popular"] = np.where(product_check["order"] > avg_orders, "Yes", "No" )

profit_check = product_check["is_profitable"] = np.where(product_check["profit"] > avg_profit, "Yes", "No")

product_check["Match?"] = np.where(
    (product_check["is_popular"] == "Yes") & (product_check["is_profitable"] == "Yes"),
    "Both", "Mismatch"
)
product_check
```
